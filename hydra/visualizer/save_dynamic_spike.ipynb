```python
init1 = 0
init2 = 123
stims_fast = [init1+0.0, init1+9.25, init1+14.5, init1+18.25, init1+21.5, init1+24.5, init1+28.0, init1+31.5, init1+35.25, init1+40.5,
              init2+0.0, init2+7.75, init2+15.25, init2+19.75, init2+23.0, init2+27.25, init2+30.5, init2+34.25, init2+38.25, init2+43.5]
stims_slow = [init1+15, init2+15]
```

```python
DPI = 200
fig = plt.figure(figsize=(1280/DPI, 660/DPI), dpi=DPI)

ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

for t in np.arange(0, 251, 1):
    
    # plt.clf()

    for j, moment in enumerate(stims_fast):
        if moment <= t:
            ax2.vlines(moment, 0, 1, color='b', linewidth=1)

    for j, moment in enumerate(stims_slow):
        if moment <= t:
            ax2.vlines(moment, 0, 1, color='r', alpha=0.5, linewidth=4)
    #         if 0 <= t - moment <= 4:
    #             plt.vlines(moment, 0, 1, color='r', alpha=0.5, linewidth=t - moment)

    ax1.axis('off')
    ax2.set_xlim(-5, 250)
    ax2.set_ylim(0, 1)
    ax2.set_yticks([])
    ax2.set_xlabel('time(s)')
    #     plt.title('Stimulation moments of fast and slow pathways')

    plt.savefig('./data/spikes_658838/img' + str(t) + '.jpg', dpi=DPI)
```
