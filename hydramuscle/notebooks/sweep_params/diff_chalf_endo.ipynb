```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import hydramuscle.utils.utils as utils
```

```python
lengths_1 = utils.length_of_model("../../results/data/length/coords_chalf_endo_001.txt", display=False)
lengths_2 = utils.length_of_model("../../results/data/length/coords_chalf_endo_01.txt", display=False)
lengths_4 = utils.length_of_model("../../results/data/length/coords_chalf_endo_02.txt", display=False)
lengths_5 = utils.length_of_model("../../results/data/length/coords_chalf_endo_05.txt", display=False)
lengths_6 = utils.length_of_model("../../results/data/length/coords_chalf_endo_1.txt", display=False)
```

```python
start = 0
end = 990

plt.figure(figsize=(10,5))
time_axis = np.arange(start, end * 0.1, 0.1)
plt.plot(time_axis, lengths_1[start : end], linewidth=3, label='$c_{half}$(endo)=0.01')
plt.plot(time_axis, lengths_2[start : end], linewidth=3, label='$c_{half}$(endo)=0.1')
plt.plot(time_axis, lengths_4[start : end], linewidth=3, label='$c_{half}$(endo)=0.2')
plt.plot(time_axis, lengths_5[start : end], linewidth=3, label='$c_{half}$(endo)=0.5')
plt.plot(time_axis, lengths_6[start : end], linewidth=3, label='$c_{half}$(endo)=1.0')
plt.legend()
plt.grid()
plt.xlabel('time(s)')
plt.ylabel('length(mm)')
plt.savefig('../../results/figures/sweep_params_chalf_endo.pdf', bbox_inches='tight')
plt.show()
```
