```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

import hydramuscle.utils.utils as utils
```

```python
lengths_1 = utils.length_of_model("../../results/data/length/coords_kf_endo_005.txt", display=False)
lengths_2 = utils.length_of_model("../../results/data/length/coords_kf_endo_01.txt", display=False)
lengths_3 = utils.length_of_model("../../results/data/length/coords_kf_endo_02.txt", display=False)
lengths_4 = utils.length_of_model("../../results/data/length/coords_kf_endo_05.txt", display=False)
lengths_5 = utils.length_of_model("../../results/data/length/coords_kf_endo_1.txt", display=False)
```

```python
start = 0
end = 990

plt.figure(figsize=(10,5))
time_axis = np.arange(start, end * 0.1, 0.1)
plt.plot(time_axis, lengths_1[start : end], linewidth=3, label='$K_F$(endo)=0.005')
plt.plot(time_axis, lengths_2[start : end], linewidth=3, label='$K_F$(endo)=0.01')
plt.plot(time_axis, lengths_3[start : end], linewidth=3, label='$K_F$(endo)=0.02')
plt.plot(time_axis, lengths_4[start : end], linewidth=3, label='$K_F$(endo)=0.05')
plt.plot(time_axis, lengths_5[start : end], linewidth=3, label='$K_F$(endo)=0.1')
plt.grid()
plt.legend()
plt.xlabel('time(s)')
plt.ylabel('length(mm)')
plt.savefig('../../results/figures/sweep_params_kf_endo.pdf', bbox_inches='tight')
plt.show()
```
