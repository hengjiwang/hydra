```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

import hydramuscle.utils.utils as utils
```

```python
lengths_1 = utils.length_of_model("../../results/data/length/coords_rlx_time_10.txt", display=False)
lengths_2 = utils.length_of_model("../../results/data/length/coords_rlx_time_100.txt", display=False)
```

```python
start = 0
end = 990

plt.figure(figsize=(10,5))
time_axis = np.arange(start, end * 0.1, 0.1)
plt.plot(time_axis, lengths_1[start:end], linewidth=3, label='relax_time=10s')
plt.plot(time_axis, lengths_2[start:end], linewidth=3, label='relax_time=100s')
plt.legend()
plt.grid()
plt.xlabel('time(s)')
plt.ylabel('length(mm)')
plt.show()
```
