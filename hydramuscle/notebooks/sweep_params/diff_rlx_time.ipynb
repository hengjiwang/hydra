```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

import hydramuscle.utils.utils as utils
```

```python
lengths_1 = utils.length_of_model("../../results/data/length/coords_rlx_time_10.txt", display=False)
lengths_2 = utils.length_of_model("../../results/data/length/coords_rlx_time_100.txt", display=False)
lengths_3 = utils.length_of_model("../../results/data/length/coords_rlx_time_200.txt", display=False)
lengths_4 = utils.length_of_model("../../results/data/length/coords_rlx_time_500.txt", display=False)
lengths_5 = utils.length_of_model("../../results/data/length/coords_rlx_time_1000.txt", display=False)
lengths_6 = utils.length_of_model("../../results/data/length/coords_rlx_time_50.txt", display=False)
```

```python
start = 0
end = 990

plt.figure(figsize=(10,5))
time_axis = np.arange(start, end * 0.1, 0.1)
plt.plot(time_axis, lengths_1[start:end], linewidth=3, label='relax_time=10s')
plt.plot(time_axis, lengths_2[start:end], linewidth=3, label='relax_time=100s')
plt.plot(time_axis, lengths_3[start:end], linewidth=3, label='relax_time=200s')
plt.plot(time_axis, lengths_4[start:end], linewidth=3, label='relax_time=500s')
plt.plot(time_axis, lengths_5[start:end], linewidth=3, label='relax_time=1000s')
plt.plot(time_axis, lengths_6[start:end], linewidth=3, label='relax_time=50s')
plt.legend()
plt.grid()
plt.xlabel('time(s)')
plt.ylabel('length(mm)')
plt.show()
```
